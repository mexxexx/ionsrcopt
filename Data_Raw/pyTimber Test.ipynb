```python
import pytimber
ldb = pytimber.LoggingDB()
```

```python
t1 = '2018-11-01 01:00:00.000'
t2 = '2018-11-02 01:00:00.000'
d = ldb.get('ITF.BCT25:CURRENT', t1, t2, unixtime=False)
print(d)
```

```python
ldb.search('ITF.BCT25:CURRENT')
```

```python
from datetime import datetime
datetime.fromtimestamp(d['ITF.BCT25:CURRENT'][0][0])
```

Connecting to pytimber works without a problem, the only thing to keep in mind that the timestamps do not coincide with the ones from Timber, but are GMT+01. That means that data that is requested in Timber from e.g. 01/11/2018 00:00 has to be requested in pytimber as 01/11/2018 01:00.

Data is returned in the format _{ Key : (timestamp array, values array) }_
